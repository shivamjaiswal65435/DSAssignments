```python
userbook = book.pivot_table( index ='User.ID',columns ='Book.Title',values ='Book.Rating')
```

```python
userbook
```

```python
userbook.fillna(0,inplace=True)
```

```python
userbook
```

```python
sim = 1-pairwise_distances(userbook.values,metric='cosine')
```

```python
sim
```

```python
np.fill_diagonal(sim,0)
```

```python
usersim = pd.DataFrame(sim)
```
